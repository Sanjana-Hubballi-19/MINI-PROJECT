```python
import pandas as pd
import os

directory_path = 'E:\MML\HUMAN_ACTIVITY_RECOGNITION\Datasets'

# Listing all CSV files in the specified directory
csv_files = [file for file in os.listdir(directory_path) if file.endswith('.csv')]

# Initializing an empty DataFrame to store the combined data
combined_data = pd.DataFrame()

# Loop through each CSV file and concatenate the data
for csv_file in csv_files:
    file_path = os.path.join(directory_path, csv_file)
    current_data = pd.read_csv(file_path)
    combined_data = pd.concat([combined_data, current_data], ignore_index=True)

combined_data.to_csv('E:\MML\HUMAN_ACTIVITY_RECOGNITION\combined_data.csv', index=False)
print(combined_data)
```

```python
import pandas as pd
from imblearn.over_sampling import SMOTE

# Load the train_data
train_data = pd.read_csv('E:\\MML\\HUMAN_ACTIVITY_RECOGNITION\\train_data.csv')

# Assuming 'timestamp' column is present, you can convert it to timestamp_seconds
#train_data['timestamp'] = pd.to_datetime(train_data['timestamp'])
#train_data['timestamp_seconds'] = train_data['timestamp'].apply(lambda x: x.timestamp())

# Define features and target variable
X_columns = ['timestamp_seconds','back_x', 'back_y', 'back_z', 'thigh_x', 'thigh_y', 'thigh_z']
X_train = train_data[X_columns]
y_train = train_data['label']

# Instantiate the SMOTE object and apply it to the training set only
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

# Create a DataFrame with the resampled training data
resampled_traindata = pd.DataFrame(X_resampled, columns=X_columns)
resampled_traindata['label'] = y_resampled

print("Shape of original training data:", X_train.shape)
print("Shape of resampled training data:", resampled_traindata.shape)
```

```python
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# Load train_data and test_data
resampled_traindata = pd.read_csv('E:\\MML\\HUMAN_ACTIVITY_RECOGNITION\\resampled_traindata.csv')
test_data = pd.read_csv('E:\\MML\\HUMAN_ACTIVITY_RECOGNITION\\test_data.csv')

X_train = resampled_traindata.drop(['label', 'timestamp_seconds'], axis=1)
y_train = resampled_traindata['label']

X_test = test_data.drop(['label', 'timestamp_seconds'], axis=1)
y_test = test_data['label']

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(1, activation='linear')
])

model.compile(optimizer='adam', loss='hinge', metrics=['accuracy'])

model.fit(X_train_scaled, y_train, epochs=10, batch_size=32, validation_data=(X_test_scaled, y_test))

y_pred = model.predict(X_test_scaled)
y_pred_binary = (y_pred > 0).astype(int)
test_acc = accuracy_score(y_test, y_pred_binary)
print(f'Test accuracy: {test_acc}')
```

```python
import pandas as pd
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
from sklearn.svm import SVC  
import matplotlib.pyplot as plt

# Load resampled training data
resampled_traindata = pd.read_csv('E:\\MML\\HUMAN_ACTIVITY_RECOGNITION\\resampled_traindata.csv')
resampled_X_train = resampled_traindata.drop('label', axis=1) 
resampled_y_train = resampled_traindata['label']

# Load test data
test_data = pd.read_csv('E:\\MML\\HUMAN_ACTIVITY_RECOGNITION\\test_data.csv')
X_test = test_data.drop('label', axis=1)  
y_test = test_data['label']


model = SVC(kernel='linear', C=1.0)  
model.fit(resampled_X_train, resampled_y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Display confusion matrix using ConfusionMatrixDisplay
cm_display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
cm_display.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()

# Display classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```
